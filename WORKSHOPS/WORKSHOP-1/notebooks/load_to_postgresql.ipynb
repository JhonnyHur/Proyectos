```python
# Conexión MySQL (origen)
mysql_url = "mysql+mysqlconnector://root:etl25@localhost:3306/workshop1_raw_db"
mysql_engine = create_engine(mysql_url)

# Conexión Postgres (destino)
pg_url = "postgresql+psycopg2://transformed:etl25@localhost:5432/workshop1_transformed_db"
pg_engine = create_engine(pg_url)

# Leer datos desde MySQL
df = pd.read_sql("SELECT * FROM raw_candidates", con=mysql_engine)

# Pasar a Postgres (nueva tabla transformed_candidates)
df.to_sql("transformed_candidates", con=pg_engine, if_exists="replace", index=False)

print("✅ Datos copiados de MySQL → Postgres:", len(df))
```
