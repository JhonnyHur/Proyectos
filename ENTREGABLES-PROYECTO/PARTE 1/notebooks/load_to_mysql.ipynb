```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import text
```

```python
user     = "root"
password  = "etl25"
host        = "localhost"
port      = "3306"
database  = "icfes_raw"

# Crear conexión SQLAlchemy (driver: mysql-connector)
url = f"mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}"
engine = create_engine(url)
```

```python
FILES = [
    "../data_raw/Examen_Saber_11_20241.txt",
    "../data_raw/Examen_Saber_11_20242.txt",
]

for file in FILES:
    print(f"\nCargando: {file}")
    df = pd.read_csv(file, sep=";", dtype=str, encoding="utf-8")

    df.to_sql(
        "examen_saber_11",
        con=engine,
        if_exists="append",
        index=False,
        chunksize=10000,
        method="multi"
    )

    print(f"--- Insertados {len(df)} registros desde {file}")

print("\n---Proceso completado la inserción de datos en MySQL ha finalizado. ")
```
